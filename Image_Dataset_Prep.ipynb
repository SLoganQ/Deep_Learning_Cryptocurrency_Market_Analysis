```python
# Image set total
num_images = 3300
```

```python
# Crop images to focus on the important data only
def crop_to_data(img_dir, num_images):
    for i in range(num_images):
        img_filename = "image_" + str(i + 1)
        img_filepath = img_dir + "/" + img_filename + ".png"
        img = Image.open(img_filepath, mode = "r")
        box = (120, 250, 2420, 1490)
        crop = img.crop(box)
        crop.save(img_filepath)


crop_to_data(img_dir, num_images)
```

```python
# Crop prediction target image into the data directory
def crop_to_pred(img_dir, pred_dir, num_images):
    for i in range(1, num_images):
        process_count = i + 1
        img_filename = "image_" + str(i + 1)
        img_filepath = img_dir + "/" + img_filename + ".png"
        new_filename = "image_" + str(i)
        pred_filepath = pred_dir + "/" + new_filename + ".png"
        img = Image.open(img_filepath, mode = "r")
        box = (1967, 0, 2300, 1240)
        crop = img.crop(box)
        img_res = crop.save(pred_filepath)
        if process_count % 100 == 0:
            print("Processed " + str(process_count) + " images so far...")
    print("Processed " + str(process_count) + ", job done.")

crop_to_pred(img_dir, pred_dir, num_images)
```

```python
# Classify prediction target images
green_pixel = (34, 158, 77)
red_pixel = (221, 68, 77)


def classify_targets(pred_dir, num_images):
    classification_list = []
    assigned = 0 # For error checking
    for i in range(num_images-1):
        process_count = i + 1
        img_filename = "image_" + str(i + 1)
        img_filepath = pred_dir + "/" + img_filename + ".png"
        pred_image = Image.open(img_filepath, mode="r")
        pred_image_filename = "image_" + str(i)
        green = 0
        red = 0
        pixels = pred_image.getdata()
        for pixel in pixels:
            if (
                pixel[0] == green_pixel[0]
                and pixel[1] == green_pixel[1]
                and pixel[2] == green_pixel[2]
            ):
                green += 1
            elif (
                pixel[0] == red_pixel[0]
                and pixel[1] == red_pixel[1]
                and pixel[2] == red_pixel[2]
            ):
                red += 1
        
        if red == 0:  # To prevent dividing by 0
            red = 1
            if green == 0:
                green == 1
                print("Error, both red and green were zero") # For error checking
        ratio = green / red
        if ratio >= 2.0:
            classification_list.append([img_filename, "gain"])
            assigned += 1
        elif ratio <= 0.5:
            classification_list.append([img_filename, "loss"])
            assigned += 1
        else:
            classification_list.append([img_filename, "neutral"])
            assigned += 1
        if process_count % 100 == 0:
            print("Processed " + str(process_count) + " images so far...")
    print("Processed " + str(process_count) + ", job done.")
    print("Assigned count: " + str(assigned))
    return classification_list


prediction_classifications = classify_targets(pred_dir, num_images)
print("Prediction_classifications_length = " + str(len(prediction_classifications)))

gains = 0
neutrals = 0
losses = 0
assigned1 = 0 # For error checking
for classification in prediction_classifications:
    if classification[1] == "gain":
        gains += 1
        assigned1 += 1
    elif classification[1] == "loss":
        losses += 1
        assigned1 += 1
    else:
        neutrals += 1
        assigned1 += 1
print(
    "Gains: " + str(gains) + ", Neutrals: " + str(neutrals) + ", Losses: " + str(losses) + ", Assigned: " + str(assigned1))
```

```python
# Resize images
reduction_factor = (
    5  # Choosing 3 is going to reduct orginal dimension to = input_dim / 5
)
input_dim = (2300, 1240)
output_dim = (
    int(input_dim[0] / reduction_factor),
    int(input_dim[1] / reduction_factor),
)


def resize_images(img_dir, num_images, ouput_dim):
    for i in range(num_images):
        process_count = i + 1
        img_filename = "image_" + str(i + 1)
        img_filepath = img_dir + "/" + img_filename + ".png"
        img = Image.open(img_filepath, mode="r")
        resized_img = img.resize(output_dim)
        img_res = resized_img.save(img_filepath)
        if process_count % 100 == 0:
            print("Processed " + str(process_count) + " images so far...")
    print("Processed " + str(process_count) + ", job done.")


resize_images(img_dir, num_images, output_dim)
```

```python
# Split image set randomly into training and validation (test) sets
def train_test_split(img_dir, training_dir, validation_dir, num_images):
    set_total = 0
    training_sum = 0
    validation_sum = 0
    for i in range(num_images):
        img_filename = "image_" + str(i + 1)
        img_filepath = img_dir + "/" + img_filename + ".png"

        random_num = random.randint(1, 101)

        # Assigns 15% of the images to the validation directory
        # Assigns 85% of the images to the training directory
        if random_num <= 85:
            # Move a file by renaming it's path
            new_path = training_dir + "/" + img_filename + ".png"
            os.rename(
                img_filepath,
                new_path,
            )
            training_sum += 1
        else:
            # Move a file by renaming it's path
            new_path = validation_dir + "/" + img_filename + ".png"
            os.rename(
                img_filepath,
                new_path,
            )
            validation_sum += 1

    percent_training = (training_sum / (training_sum + validation_sum)) * 100
    percent_validation = (validation_sum / (training_sum + validation_sum)) * 100
    print(
        "(Number of training images, percent of set): ("
        + str(training_sum)
        + ", "
        + str(percent_training)
        + ")"
    )
    print(
        "(Number of validation images, percent of set): ("
        + str(validation_sum)
        + ", "
        + str(percent_validation)
        + ")"
    )


train_test_split(img_dir, training_data_dir, validation_data_dir, num_images)
```

```python
# Go through the training and validation images and use the classifications
# to assign each image to their respective class directory
def assign_to_class_dir(
    main_dir, sample_size, gain_dir, neutral_dir, loss_dir, classifications
):
    img_filenames = [f for f in os.listdir(main_dir) if isfile(join(main_dir, f))]
    for img_filename in img_filenames:
        for classification in classifications:
            classification_filename = classification[0] + ".png"
            if classification_filename == img_filename:
                if classification[1] == "gain":
                    image_filepath = main_dir + "/" + img_filename
                    new_path = gain_dir + "/" + img_filename
                    os.rename(image_filepath, new_path)
                elif classification[1] == "loss":
                    image_filepath = main_dir + "/" + img_filename
                    new_path = loss_dir + "/" + img_filename
                    os.rename(image_filepath, new_path)
                else:
                    image_filepath = main_dir + "/" + img_filename
                    new_path = neutral_dir + "/" + img_filename
                    os.rename(image_filepath, new_path)


assign_to_class_dir(
    training_data_dir,
    training_sample_size,
    gain_training_data_dir,
    neutral_training_data_dir,
    loss_training_data_dir,
    prediction_classifications,
)
assign_to_class_dir(
    validation_data_dir,
    validation_sample_size,
    gain_validation_data_dir,
    neutral_validation_data_dir,
    loss_validation_data_dir,
    prediction_classifications,
)
```
